```python
import pandas as pd
import numpy as np
from selenium import webdriver
from bs4 import BeautifulSoup
from tqdm import tqdm_notebook
from urllib.request import urlopen

import time
import re
```

```python
driver = webdriver.Chrome('chromedriver.exe')

url = 'https://www.hyundaicard.com/cpu/cs/CPUCS0701_01.hc'
driver.get(url)

for i in range(3):
    driver.find_element_by_xpath('//span[@id="moreNum"]').click()
```

```python
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
title = soup.select('p.location')
Title = []

for i in range(len(title)) :
    tmp_t = title[i].text.strip()
    
    Title.append(tmp_t)
```

```python
df1_t = pd.DataFrame(Title)
```

```python
df1_t
```

```python
df1=pd.DataFrame(df1_t[0].str.split(':').str[3])
```

```python
raw_data={'1':df1[0].str.split(' ').str[1],
          '2':df1[0].str.split(' ').str[2],
          '3':df1[0].str.split(' ').str[3]}

data = pd.DataFrame(raw_data)
data
```

```python
data.to_csv("현대카드.csv", index=False, encoding='CP949')
```
