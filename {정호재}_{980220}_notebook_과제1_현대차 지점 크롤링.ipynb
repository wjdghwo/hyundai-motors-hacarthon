```python
from bs4 import BeautifulSoup
from selenium import webdriver
from time import sleep

import pandas as pd
```

```python
driver = webdriver.Chrome('chromedriver.exe')
sleep(3)

driver.get('https://www.hyundai.com/kr/ko/e/customer/branch')
sleep(3)
```

```python
final_df = pd.DataFrame()
```

```python
pageString = driver.page_source
bs = BeautifulSoup(pageString, 'html.parser')
title = bs.select('b')
location=bs.find_all('span')[285:400]

store_title=[]
store_location=[]

for i in range(10):
    store_title.append(title[i].text.strip())
    store_location.append(location[4*i+2].text.strip())
```

```python
new_data = pd.DataFrame(data = {'업체명' : store_title, '주소' : store_location})
final_df = pd.concat([final_df, new_data])
final_df
```

```python
final_df.to_csv("타지역.csv", index=False, encoding='CP949')
```
